## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

### Cleaning Data

```python
combined_df = pd.concat(dataframes.values(), ignore_index=True)

print("5 baris pertama dari combined_df:")
print(combined_df.head())
print(f"\nBentuk combined_df: {combined_df.shape}")
```

```python
if 'year' in combined_df.columns and 'month' in combined_df.columns and 'day' in combined_df.columns and 'hour' in combined_df.columns:
    combined_df['datetime'] = pd.to_datetime(combined_df[['year', 'month', 'day', 'hour']])
    combined_df = combined_df.drop(columns=['year', 'month', 'day', 'hour'])
    print("  -> Kolom 'datetime' berhasil dibuat dan kolom waktu asli dihapus.")
else:
    print("  -> Kolom tahun, bulan, hari, atau jam tidak ditemukan. Kolom 'datetime' diasumsikan sudah ada.")
```

```python
# Membuat kolom turunan untuk 'hari dalam seminggu' dan 'musim'
if 'datetime' in combined_df.columns:
    if 'day_of_week' not in combined_df.columns:
        combined_df['day_of_week'] = combined_df['datetime'].dt.day_name()
        print("  -> Kolom 'day_of_week' berhasil dibuat.")
    else:
        print("  -> Kolom 'day_of_week' sudah ada.")

    if 'is_weekday' not in combined_df.columns:
        combined_df['is_weekday'] = combined_df['datetime'].dt.dayofweek < 5
        print("  -> Kolom 'is_weekday' berhasil dibuat.")
    else:
        print("  -> Kolom 'is_weekday' sudah ada.")

    def get_season(month):
        if month in [12, 1, 2]:
            return 'Winter'
        elif month in [3, 4, 5]:
            return 'Spring'
        elif month in [6, 7, 8]:
            return 'Summer'
        else:
            return 'Autumn'

    if 'season' not in combined_df.columns:
        combined_df['season'] = combined_df['datetime'].dt.month.apply(get_season)
        print("  -> Kolom 'season' berhasil dibuat.")
    else:
        print("  -> Kolom 'season' sudah ada.")
else:
    print("  -> Kolom 'datetime' tidak ditemukan, melewatkan pembuatan kolom turunan waktu.")
```

```python
# Mengidentifikasi dan mengklasifikasikan setiap stasiun sebagai 'pusat kota' atau 'pinggiran'
if 'station' in combined_df.columns and 'location_type' not in combined_df.columns:
    suburban_stations = ['Changping', 'Dingling', 'Huairou', 'Shunyi', 'Wanliu']
    city_center_stations = ['Aotizhongxin', 'Dongsi', 'Guanyuan', 'Gucheng', 'Nongzhanguan', 'Tiantan', 'Wanshouxigong']

    def classify_station_location(station_name):
        if station_name in suburban_stations:
            return 'Suburban'
        elif station_name in city_center_stations:
            return 'City Center'
        else:
            return 'Unknown'

    combined_df['location_type'] = combined_df['station'].apply(classify_station_location)
    print("  -> Kolom 'location_type' untuk stasiun berhasil dibuat.")
else:
    print("  -> Kolom 'location_type' sudah ada atau kolom 'station' tidak ditemukan.")
```

#### Fixing lost value

```python
# Menangani nilai yang hilang (missing values)
print("\n--- Menangani Nilai yang Hilang ---")
print("Jumlah nilai yang hilang sebelum imputasi:")
initial_missing = combined_df.isnull().sum()[combined_df.isnull().sum() > 0]
if not initial_missing.empty:
    print(initial_missing)
else:
    print("Tidak ada nilai yang hilang sebelum imputasi.")

numerical_cols = combined_df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = combined_df.select_dtypes(include='object').columns.tolist()

# Hapus kolom 'No' dari daftar numerik untuk imputasi jika ada
if 'No' in numerical_cols:
    numerical_cols.remove('No')

print("\n  -> Melakukan imputasi kolom numerik dengan median...")
for col in numerical_cols:
    if combined_df[col].isnull().any():
        median_val = combined_df[col].median()
        combined_df[col].fillna(median_val, inplace=True)
        print(f"    Kolom '{col}' diimputasi dengan median: {median_val}")

print("\n  -> Melakukan imputasi kolom kategorikal dengan modus...")
for col in categorical_cols:
    if combined_df[col].isnull().any():
        # Untuk 'wd' yang memiliki missing values, kita isi dengan 'Unknown' agar tidak bias pada modus
        if col == 'wd':
            combined_df[col].fillna('Unknown', inplace=True)
            print(f"    Kolom '{col}' diimputasi dengan 'Unknown'.")
        else:
            mode_val = combined_df[col].mode()[0]
            combined_df[col].fillna(mode_val, inplace=True)
            print(f"    Kolom '{col}' diimputasi dengan modus: {mode_val}")
```

```python
print("\nJumlah nilai yang hilang setelah imputasi:")
missing_after_imputation = combined_df.isnull().sum()[combined_df.isnull().sum() > 0]
if missing_after_imputation.empty:
    print("Tidak ada nilai yang hilang setelah imputasi.")
else:
    print(missing_after_imputation)

print("\n--- Ringkasan DataFrame setelah Preprocessing dan Cleaning ---")
combined_df.info()
print("\n5 baris pertama dari combined_df setelah semua preprocessing:")
print(combined_df.head())
```

### Pertanyaan 1: Apakah rata-rata konsentrasi emisi gas buang kendaraan (NO2 dan CO) pada jam sibuk di stasiun pusat kota secara presentasi lebih tiggi dibandingkan dengan stasiun area pinggiran?

```python
# 1. Mendefinisikan 'jam sibuk' (07:00-09:00 dan 16:00-18:00).
rush_hour_morning = (combined_df['datetime'].dt.hour >= 7) & (combined_df['datetime'].dt.hour < 10)
rush_hour_evening = (combined_df['datetime'].dt.hour >= 16) & (combined_df['datetime'].dt.hour < 19)
```

```python
# 2. Memfilter data hanya untuk jam sibuk.
rush_hour_df = combined_df[rush_hour_morning | rush_hour_evening].copy()
print(f"Data jam sibuk berhasil difilter. Jumlah data: {len(rush_hour_df)} baris.")
```

```python
# 3. Mengelompokkan data berdasarkan jenis stasiun (pusat kota vs pinggiran).
# 4. Menghitung rata-rata konsentrasi NO2 dan CO untuk masing-masing kelompok.
average_emissions = rush_hour_df.groupby('location_type')[['NO2', 'CO']].mean()
print("\nRata-rata konsentrasi NO2 dan CO selama jam sibuk berdasarkan tipe lokasi:")
print(average_emissions)

# Memisahkan rata-rata untuk pusat kota dan pinggiran
if 'City Center' in average_emissions.index and 'Suburban' in average_emissions.index:
    city_center_avg = average_emissions.loc['City Center']
    suburban_avg = average_emissions.loc['Suburban']

    print("\n--- Perbandingan Rata-rata Emisi ---")
    print(f"Rata-rata NO2 di Pusat Kota: {city_center_avg['NO2']:.2f}")
    print(f"Rata-rata NO2 di Pinggiran: {suburban_avg['NO2']:.2f}")
    print(f"Rata-rata CO di Pusat Kota: {city_center_avg['CO']:.2f}")
    print(f"Rata-rata CO di Pinggiran: {suburban_avg['CO']:.2f}")
```

```python
# 5. Menghitung perbandingan persentase antara rata-rata konsentrasi di stasiun pusat kota dan pinggiran.

if 'City Center' in average_emissions.index and 'Suburban' in average_emissions.index:
    # Asumsi: Menghitung persentase perbedaan relatif terhadap area pinggiran
    percentage_diff_no2 = ((city_center_avg['NO2'] - suburban_avg['NO2']) / suburban_avg['NO2']) * 100
    percentage_diff_co = ((city_center_avg['CO'] - suburban_avg['CO']) / suburban_avg['CO']) * 100

    print("\n--- Perbandingan Persentase (Pusat Kota vs Pinggiran) ---")
    print(f"Konsentrasi NO2 di Pusat Kota {'lebih tinggi' if percentage_diff_no2 > 0 else 'lebih rendah'} {abs(percentage_diff_no2):.2f}% dibandingkan Pinggiran.")
    print(f"Konsentrasi CO di Pusat Kota {'lebih tinggi' if percentage_diff_co > 0 else 'lebih rendah'} {abs(percentage_diff_co):.2f}% dibandingkan Pinggiran.")

    # Menentukan apakah secara presentase lebih tinggi
    if percentage_diff_no2 > 0 and percentage_diff_co > 0:
        print("\nKesimpulan Sementara: Rata-rata konsentrasi emisi NO2 dan CO di stasiun pusat kota secara presentasi lebih tinggi dibandingkan dengan stasiun area pinggiran.")
    elif percentage_diff_no2 < 0 and percentage_diff_co < 0:
        print("\nKesimpulan Sementara: Rata-rata konsentrasi emisi NO2 dan CO di stasiun pusat kota secara presentasi lebih rendah dibandingkan dengan stasiun area pinggiran.")
    else:
        print("\nKesimpulan Sementara: Perbandingan konsentrasi emisi NO2 dan CO di stasiun pusat kota dan pinggiran menunjukkan pola yang bervariasi.")
else:
    print("\nTidak dapat melakukan perbandingan persentase karena data untuk 'City Center' atau 'Suburban' tidak lengkap.")
```

**Insight:**

**Rata-rata Emisi:** Selama jam sibuk (07:00-09:00 dan 16:00-18:00), rata-rata konsentrasi NO2 dan CO lebih tinggi di stasiun Pusat Kota dibandingkan stasiun Pinggiran.
    *   Rata-rata NO2 di Pusat Kota: **52,40 µg/m³**
    *   Rata-rata NO2 di Pinggiran: **42,41 µg/m³**
    *   Rata-rata CO di Pusat Kota: **1237,72 µg/m³**
    *   Rata-rata CO di Pinggiran: **1078,60 µg/m³**
*   **Perbedaan Persentase:**
    *   Konsentrasi NO2 di Pusat Kota **23,56% lebih tinggi** dibandingkan area Pinggiran.
    *   Konsentrasi CO di Pusat Kota **14,75% lebih tinggi** dibandingkan area Pinggiran.

### Pertanyaan 2: Berapa presentasi waktu (dalam jam) dan total hari terjadinya kondisi saat konsentrasi NO2 melampaui 80 µg/m³ bersamaan dengan keceatan angin di bawah 1.5 m/s pada hari kerja selama musim dingin?

```python
# 1. Memfilter data untuk 'hari kerja' dan 'musim dingin'.
winter_weekday_df = combined_df[(combined_df['is_weekday'] == True) & (combined_df['season'] == 'Winter')].copy()
print(f"Data difilter untuk hari kerja di musim dingin. Jumlah data: {len(winter_weekday_df)} baris.")

# Menghitung total jam dan hari unik yang diamati selama hari kerja di musim dingin
total_winter_weekday_hours = len(winter_weekday_df)
total_winter_weekday_days = winter_weekday_df['datetime'].dt.date.nunique()
print(f"Total jam di hari kerja musim dingin: {total_winter_weekday_hours}")
print(f"Total hari unik di hari kerja musim dingin: {total_winter_weekday_days}")
```

```python
# 2. Mengidentifikasi data di mana konsentrasi NO2 > 80 µg/m³ DAN kecepatan angin (WSPM) < 1.5 m/s.
# Pastikan kolom NO2 dan WSPM tidak null (meskipun sudah diimputasi, ini untuk keamanan)
condition_met_df = winter_weekday_df[
    (winter_weekday_df['NO2'] > 80) &
    (winter_weekday_df['WSPM'] < 1.5)
].copy()

print(f"Data memenuhi kondisi NO2 > 80 µg/m³ dan WSPM < 1.5 m/s: {len(condition_met_df)} baris.")
```

```python
# 3. Menghitung total jam (jumlah baris) yang memenuhi kondisi ini.
total_hours_condition_met = len(condition_met_df)
print(f"Total jam yang memenuhi kondisi: {total_hours_condition_met}")
```

```python
# 4. Menghitung total hari unik yang memenuhi kondisi ini.
total_days_condition_met = condition_met_df['datetime'].dt.date.nunique()
print(f"Total hari unik yang memenuhi kondisi: {total_days_condition_met}")
```

```python
# 5. Menghitung persentase jam dan hari tersebut terhadap total jam dan hari yang diamati selama hari kerja di musim dingin.
percentage_hours = (total_hours_condition_met / total_winter_weekday_hours) * 100 if total_winter_weekday_hours > 0 else 0
percentage_days = (total_days_condition_met / total_winter_weekday_days) * 100 if total_winter_weekday_days > 0 else 0

print(f"\nPersentase waktu (dalam jam) kondisi terpenuhi: {percentage_hours:.2f}%")
print(f"Persentase total hari kondisi terpenuhi: {percentage_days:.2f}%")
```

**Insight:**

 **Data Terfilter:** Analisis berfokus pada **74,592 jam** di **259 hari unik** yang diklasifikasikan sebagai hari kerja musim dingin.
*   **Kejadian Kondisi Kritis:** Kondisi kritis spesifik di mana konsentrasi NO2 melebihi 80 µg/m³ dan kecepatan angin (WSPM) di bawah 1.5 m/s terjadi selama:
    *   **Total Jam:** 15,627 jam
    *   **Total Hari Unik:** 224 hari
*   **Persentase Kejadian:**
    *   Kondisi tersebut terjadi dalam **20.95%** dari total jam hari kerja musim dingin.
    *   Kondisi tersebut terjadi pada **86.49%** dari total hari kerja musim dingin.
